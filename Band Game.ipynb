For this project I created a 'Band' class, which creates a representation of a band of musicians. It stores members, corresponding instruments and energy levels, total funds, songs written and other related attributes.

I also included a function '$ runsim() $', which generates a text-prompt adventure game of sorts in which you guide your created band through the soul-destroying experience of living as performing musicians. 

The source code can be seen in the cell below:

```python
import random, time

class Band:
  def __init__(self, title, members, money):
    if not isinstance(members, int):
      raise ValueError('Number of members must be an integer')
    if members > 6:
      raise ValueError("That's far too many band members! How do you expect to make any kind of profit if you're splitting the money {} ways?".format(members))
    if members < 2:
      raise ValueError("That's not a band, that's a solo project. Stick to soundcloud".format(members))
    if members < 1:
      raise ValueError("No, you can't have a band consisting of less than one member".format(members))
    self.members = members
    self.money = money
    self.title = title
    
    
  
  def assignnames(self):
    self.memberlist = []
    for i in range(self.members):
      memb = input('Enter name of band member {}: '.format(i + 1))
      self.memberlist.append(memb)
    return self.memberlist


  def instruments(self, members):
    instlist = ['guitar','bass','keyboards', 'vocals','saxophone','percussion','winds']
    playerinst = []
    for i in members:
      randinst = random.randint(0, len(instlist)-1)
      playerinst.append(instlist[randinst])
      instlist.remove(instlist[randinst])
    return playerinst

  def chooseinst(self, members):
    instlist = ['guitar','bass','keyboards', 'vocals','saxophone','percussion','winds']
    print('Choose from the following list of instruments for each band member:')
    for i in range(len(instlist)):
      print(i,': ',instlist[i])
    playerinst = []
    print('Input the number for the desired instrument for each member:')
    for i in range(len(members)):
      print(members[i],': ')
      ch = eval(input('   '))
      if not isinstance(ch, int) or ch > (len(instlist) - 1) or ch < 0:
        raise ValueError('Invalid value')
      inst = instlist[ch]
      playerinst.append(inst)
    return playerinst


  def energylvlset(self, members):
    energylvls = []
    for i in members:
      energylvls.append(100.0)
    return energylvls
    
  def writesong(self, member):
    songtitle = input('{} is writing a song, what will they call it? '.format(member))
    return songtitle

  def playgig(self, title, members, instruments, songs, money, energy):
    earnings = 0
    yn = input('A venue has offered the band a gig tonight! Will they take it? y/n: ')

    if yn == 'y':
      setlist = []
      for i in range(5):
        randsong = random.randint(0,len(songs)-1)
        setlist.append(songs[randsong])
        songs.remove(songs[randsong])
      print('The band takes the gig, with the following setlist prepared:')
      print(setlist)
      time.sleep(1)
      for i in setlist:
        print('The band starts playing {}'.format(i))
        time.sleep(1)
        rand = random.randint(0,len(members)-1)
        randmemb = members[rand]
        randinstr = instruments[rand]
        yn = input('Take a solo? y/n: ')
        if yn == 'y':
          print('{} takes a {} solo midway through the song'.format(randmemb,randinstr))
          time.sleep(1)
          if energy[rand] == 100:
            print("The solo is jaw-dropping, all eyes are on {}".format(randmemb))
            earnings = earnings + 10
          elif energy[rand] >= 50:
            print("The audience seem to like it")
            earnings = earnings + 5
          elif energy[rand] < 50 and energy[rand] > 20:
            print("Maybe a solo wasn't such a good idea, the audience seem kind of bored")
          else:
            print("The solo is comically awful")
            earnings = earnings - 5
          energy[rand] = energy[rand] - 20
        for i in range(len(energy)):
          energy[i] = energy[i] - 15
          if energy[i] < 0:
            energy[i] = 0
        earnings = earnings + 25
        time.sleep(2)
        print('The song ends')
        print('')
      print('The gig is over, the venue payed you €{}'.format(earnings))
      return earnings

    else: 
      print('The band decide to give it a miss')
      return earnings

  
  def albumdeal(self, title, members, instruments, songs, money, energy):
    albumtitle = input("It's finally happened! Mike Hawk from E&I Records has decided to sign the band to his label! They're hoping to release an album of theirs in two months time. What will they call the album?")
    print('Time to start work on ',albumtitle,' then!')
    time.sleep(1)
    trackno = eval(input('How many tracks will they put on the album?'))
    
    print('What songs will they put on the album?')
    for i in range(len(songs)):
      print(i,': ',songs[i])
    print('Input the number for the desired song:')
    tracklist = []
    for i in range(trackno):
      print('Track ',i + 1,': ')
      ch = eval(input('   '))
      if not isinstance(ch, int) or ch > (len(songs) - 1) or ch < 0:
        raise ValueError('Invalid value')
      track = songs[ch]
      tracklist.append(track)
    print('Your album will be released today!')
    print(albumtitle,':')
    for i in range(len(tracklist)):
      print(i + 1,': ',tracklist[i])
    return albumtitle
    


  def runsim(self, choose=False):
    print('Starting Simulation')
    time.sleep(0.7)
    funi = ['Loading Artistic Discontent','Loading Poverty','Loading Baying Masses', 'Loading Indulgent Keyboard Solos', 'Loading the Tendency to Try Out Every Guitar in the Shop then Leave After Buying a Pick','Loading Unsuccessful Follow-Up Solo Careers','Loading Unvisited RYM Pages','Loading Embarrassing and Downright Unsanitary Living Arrangements','Loading Faulty Jack-Leads','Loading Pints Emptied into Amplifiers']
    for i in range(3):
      print(funi[random.randint(0, 9)])
      time.sleep(0.7)

    print('')
    print('')
    title = self.title
    money = self.money
    members = self.assignnames()
    print('')
    if choose:
      instruments = self.chooseinst(members)
    else:
      instruments = self.instruments(members)
    energy = self.energylvlset(members)
    print('')
    print('A band has been formed called {}, consisting of:'.format(title))
    for i in range(len(members)):
      print(members[i],' on ',instruments[i])
    songs = []
    print('')
    print('')
    time.sleep(3)
    albumout = False
    while money != 0:
      time.sleep(1)
      rng = random.randint(0, 3)
      if rng == 0:
        randmemb = members[random.randint(0, len(members)-1)]
        songs.append(self.writesong(randmemb))
      elif rng == 1 and len(songs) > 4:
        earnings = self.playgig(title, members, instruments, songs, money, energy)
        if earnings == 0:
          self.energylvlset(members)
        else: 
          for i in range(len(energy)):
            energy[i] = energy[i] - 25
        money = money + earnings
      elif rng == 2:
        randmemb = random.randint(0, len(members)-1)
        randmoney = random.randint(10, 30)
        money = money + randmoney
        print(members[randmemb],' receives €',randmoney," to contribute to the band's finances")
        randmemb = members[random.randint(0, len(members)-1)]
        songs.append(self.writesong(randmemb))
      elif rng == 3 and len(songs) > 5 and albumout == False:
        albumtitle = self.albumdeal(title, members, instruments, songs, money, energy)
        albumout = True

      time.sleep(2)
      print('The band has finished work for the day')
      if albumout:
        print('Your album {} has pulled in €50 from sales today!'.format(albumtitle))
        money = money + 50
      for i in range(len(members)):
        print(members[i],' energy level: ',energy[i])
      money = money - (6 * len(members))
      if money < 0:
        money = 0
      print('Funds remaining: €',money)
      for i in range(len(energy)):
        energy[i] = energy[i] + 20
      if sum(energy) > 100:
        energy = self.energylvlset(members)
      
      print('')
      print('')
    print('{} have out of money and its members have returned to working in the asbestos mines. A dream never lasts long.'.format(title))
      
```

```python
#class definition takes 3 args: title, members and money. These dictate the band name, the number of members and the starting funds, respectively

MyBand = Band('The B Sharps', 4, 400)

#runsim takes one arg - choose. This dictates whether or not the instruments played are chosen by the user or randomly selected
#notably when choose is set to false, multiple band members can't play the same instrument, for the sake of avoiding generating a band consisting of just bassists 

MyBand.runsim(choose=True)
```
